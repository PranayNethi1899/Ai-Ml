```python
import random
def show(h,g):
    
    print(g)
    print("\n\nGuess the word\n\nLives remaining: "+str(h)+"\n")

print("Welcome to Nine Lives\n\n")

l=['brave','worst','blood','flesh','grave','shady','dirty','black','bleed','blade','blast']
w=random.choice(l)
h=9

g=['?','?','?','?','?']
while(1):
    show(h,g)
    a=input("Enter a letter or a word: ")
    if len(a)==1:
        if a in w:
            x=g.count(a)
            if x==w.count(a):
                h=h-1
            else:
                p=0
                for i in range(0,len(w)):
                    if w[i]==a:
                        p=p+1
                    if p>x:
                        break
                g[i]=a
                s=""
                for k in g:
                    s=s+k
                if s==w:
                    print("\n\n\nYou Win\nWord is: "+w+"\n\n")
                    break
            
        else:
            h=h-1
    else:
        if a==w:
            print("\n\n\nYou Win\nWord is: "+w+"\n\n")
            break
        else:
            h=h-1
    if h==0:
        print("\nLives: "+str(h)+"\n\n\nYou Loose\n\n\n")
        break
```
